Se realiza la importación de los módulos que ocuparemos para la realización del programa.

```python
import operator,csv,functools
```

Creamos una función para leer el archivo de extensión csv, quitamos los delimitadores y comenzamos a recorrer el archivo mediante un for, dentro de dicho ciclo se irán metiendo los datos a una lista declarada con anterioridad, cada elemento de esta lista será una lista para así como parte final devolver a través del return una lista de listas que contienen la información del csv leído. 

```python
def read_csv(archivo:str):
    with open(archivo) as csvfile:
        reader = csv.reader(csvfile, delimiter=',')
        next(reader)
        data =[]
        for fila in reader:
            data.append(list(fila))
    return data
```

Una vez que tenemos los datos del archivo csv, definimos una función con nombre teams_winers que recibe como parámetro una lista de listas con los datos de cada partido, esta función tendrá como finalidad obtener todos los datos de aquellos equipos que hallan ganado un partido como local, para esto realizamos un filter al cual le pasaremos los datos de todos los partidos y como función de filtrado una lambda que comparará si la posición numero 5 de la lista recibida es igual a “H”, aquellas listas que cumplan con la condición formaran parte del iterable que retorna la función filter.

```python
def teams_winers(datos):
    return filter(lambda x: x[5]=='H',datos)
```

Para la realización del punto numero 3, utilizaremos un closer que como parámetro de la primera función recibirá los datos obtenidos en el punto anterior y como parámetro de la segunda función recibiremos el nombre del equipo del que queremos conocer la sumatoria total tanto de goles recibidos como anotados, para esto en la segunda función retornaremos un diccionario, el cual llevara como primer elemento el nombre del equipo, como segundo elemento tenemos la sumatoria de todos los goles anotados, la cual calcularemos a través de la función reduce que recibe como función a operator.add y como iterable un list comprenhension que contiene los números de goles anotados en cada partido, para la creación de este list se itero sobre el resultado de un filter en el que se recibe la lista de listas con toda la información de los partidos y como función filtradora una lambda que comprara la posición numero 1 de la lista con el nombre del equipo actual que deseamos conocer la sumatoria, por ultimo para el tercer elemento del diccionario se realiza exactamente lo mismo que se menciono anteriormente solamente que con los goles en contra. 

```python
def closer(datos):
    def compa(x):
        return {'HomeTeam':x,'Total FTHG':functools.reduce(operator.add,[int(i[3]) for i in filter(lambda y: y[1]==x,datos)]),'Total FTAG':functools.reduce(operator.add,[int(i[4]) for i in filter(lambda y: y[1]==x,datos)])}
    return compa
```

La función “Principal” para el punto 3, recibe como parámetro los datos obtenidos en el anterior punto, se hace una asignación para poder ocupar el closer, desde aquí es donde le pasamos los datos para posteriormente retornar el resultado de un map que recibe como iterable un conjunto con todos los nombres de los equipos que han ganado como local y la función que se aplicara dentro del map es la segunda función del closer. 

```python
def goalsbyteam(datos):
    a = closer(datos)
    return map(a,{x[1] for x in datos})
```

Para el ultimo punto, se nos pide que mostremos los equipos que tenga un 85% de goles anotados como locales, definimos una función que recibe como parámetro los datos obtenidos en el punto anterior y con la ayuda de un filter que recibe la lista de diccionarios con el nombre y sumatorias de goles y como función filtradora una lambda la cual hace uso de la función operator.truediv para realizar el cociente, como recibe dos parámetros(el dividendo y el divisor) hacemos uso de pipelines obteniendo el dividendo del resultado de la función operator.mul, la cual recibe como primer parámetro el valor de la clave ‘Total FTHG’ y como segundo el numero entero 100, para la parte del divisor ocupamos a la función operator.add que recibe los valores de las claves ‘Total FTHG’ y ‘Total FTAG’.

El resultado obtenido del cociente se compara con 85 para saber si es mayor a este, en dado caso de que lo sea la función filter meterá al iterable al diccionario con la información del equipo.


```python
def goal_percentage(datos):
    return filter(lambda x: operator.truediv(operator.mul(x['Total FTHG'],100),operator.add(x['Total FTHG'],x['Total FTAG']))>85,datos)
```

A continuación se mostraran los resultados obtenidos en cada punto:
